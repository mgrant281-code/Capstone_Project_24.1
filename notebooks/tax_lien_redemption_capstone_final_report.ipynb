## 9. Outlier analysis

The rubric specifically asks for anomaly and outlier analysis. Since the financial variables are strongly skewed, I am not removing all outliers automatically. Instead, I want to identify them, document them, and decide whether they should be handled differently for plots versus modeling.


```python
def iqr_outlier_summary(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    mask = (series < lower) | (series > upper)
    return {
        "q1": q1,
        "q3": q3,
        "iqr": iqr,
        "lower_bound": lower,
        "upper_bound": upper,
        "outlier_count": int(mask.sum()),
        "outlier_pct": float(mask.mean())
    }

outlier_results = pd.DataFrame({
    col: iqr_outlier_summary(df_clean[col].dropna())
    for col in numeric_core
}).T

outlier_results
```

The outlier summary shows that these core financial variables contain many high-end values. For visualization, I trimmed at the 99th percentile so the plots would stay readable. For modeling, I kept the values because logistic regression can still work with them after scaling, and those extreme cases may carry meaningful repayment information.


## 10. Feature engineering

Feature engineering is one of the most important parts of this module because raw variables do not always capture financial stress very well. I created several derived features that are more meaningful for repayment risk.

These also connect directly to the tax lien project because they mimic the kind of financial stress and capacity signals I would want in a real redemption model.


```python
df_fe = df_clean.copy()

# Known anomaly in this dataset: 365243 represents missing employment days
df_fe["DAYS_EMPLOYED_ANOM"] = (df_fe["DAYS_EMPLOYED"] == 365243).astype(int)
df_fe["DAYS_EMPLOYED"] = df_fe["DAYS_EMPLOYED"].replace(365243, np.nan)

# Human-readable versions
df_fe["YEARS_BIRTH"] = -df_fe["DAYS_BIRTH"] / 365
df_fe["YEARS_EMPLOYED"] = -df_fe["DAYS_EMPLOYED"] / 365

# Financial stress ratios
df_fe["CREDIT_INCOME_RATIO"] = df_fe["AMT_CREDIT"] / df_fe["AMT_INCOME_TOTAL"]
df_fe["ANNUITY_INCOME_RATIO"] = df_fe["AMT_ANNUITY"] / df_fe["AMT_INCOME_TOTAL"]
df_fe["GOODS_CREDIT_RATIO"] = df_fe["AMT_GOODS_PRICE"] / df_fe["AMT_CREDIT"]

# Aggregate external credit signal
# The columns EXT_SOURCE_1, EXT_SOURCE_2, EXT_SOURCE_3 were dropped due to high missing values.
# Therefore, EXT_SOURCE_MEAN cannot be created.
# df_fe["EXT_SOURCE_MEAN"] = df_fe[["EXT_SOURCE_1", "EXT_SOURCE_2", "EXT_SOURCE_3"]].mean(axis=1)

feature_preview = df_fe[[
    "TARGET", "CREDIT_INCOME_RATIO", "ANNUITY_INCOME_RATIO",
    "GOODS_CREDIT_RATIO", # "EXT_SOURCE_MEAN", # Removed as its source columns were dropped
    "YEARS_BIRTH",
    "YEARS_EMPLOYED", "DAYS_EMPLOYED_ANOM"
]].head()

feature_preview
```

## 12. Baseline modeling setup

For the baseline model, I chose **logistic regression**.

### Why logistic regression?
- It is a standard baseline for binary classification.
- It is relatively easy to interpret.
- It works well with a preprocessing pipeline for mixed data types.
- It gives me a clean benchmark before trying stronger models in a later module.

### Why ROC-AUC?
I chose **ROC-AUC** as the primary evaluation metric because:

- the target is imbalanced
- I care more about ranking higher-risk vs lower-risk cases than using one fixed threshold
- ROC-AUC is a common and defensible baseline metric for this type of classification problem


```python
numeric_features = [
    "AMT_INCOME_TOTAL", "AMT_CREDIT", "AMT_ANNUITY", "AMT_GOODS_PRICE",
    "REGION_POPULATION_RELATIVE", "DAYS_BIRTH", "DAYS_EMPLOYED",
    "DAYS_REGISTRATION", "DAYS_ID_PUBLISH", "CNT_CHILDREN", "CNT_FAM_MEMBERS",
    # "EXT_SOURCE_1", "EXT_SOURCE_2", "EXT_SOURCE_3", # Dropped due to high missing values
    "YEARS_BIRTH", "YEARS_EMPLOYED",
    "CREDIT_INCOME_RATIO", "ANNUITY_INCOME_RATIO",
    "GOODS_CREDIT_RATIO", # "EXT_SOURCE_MEAN", # Removed as its source columns were dropped
    "DAYS_EMPLOYED_ANOM"
]

categorical_features = [
    "NAME_CONTRACT_TYPE", "CODE_GENDER", "FLAG_OWN_CAR", "FLAG_OWN_REALTY",
    "NAME_INCOME_TYPE", "NAME_EDUCATION_TYPE", "NAME_FAMILY_STATUS",
    "NAME_HOUSING_TYPE", "OCCUPATION_TYPE", "ORGANIZATION_TYPE"
]

X = df_fe[numeric_features + categorical_features]
y = df_fe["TARGET"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=42
)

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

baseline_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", LogisticRegression(
        max_iter=2000,
        class_weight="balanced",
        solver="liblinear"
    ))
])

baseline_model.fit(X_train, y_train)
pred_proba = baseline_model.predict_proba(X_test)[:, 1]

roc_auc = roc_auc_score(y_test, pred_proba)
precision, recall, _ = precision_recall_curve(y_test, pred_proba)
pr_auc = auc(recall, precision)

print(f"ROC-AUC: {roc_auc:.4f}")
print(f"PR-AUC: {pr_auc:.4f}")
```

The executed notebook produced a ROC-AUC of 0.6678, which indicates **the model learned meaningful signal but still has substantial room for improvement.** The biggest likely reason the ROC-AUC is lower than it could be is likely **feature loss,** which can be corrected. For an initial capstone baseline, I think this is a good result because it shows the problem is learnable even before using more advanced models.

## 17. Improved modeling plan

The baseline logistic regression produced a ROC-AUC of approximately 0.668, which shows that the model learned meaningful signal but still has room for improvement. Because this is an imbalanced classification problem, ROC-AUC is a useful ranking metric, and PR-AUC is also important because it reflects performance on the minority class more directly.

Three issues likely limited baseline performance:

1. Important predictors may have been excluded, especially the `EXT_SOURCE_*` variables, which are often highly informative in this dataset.
2. A single train/test split can produce unstable estimates, so stratified cross-validation is needed.
3. A single untuned model may underperform compared with tuned logistic regression and stronger tree-based methods.

To improve the modeling workflow, the next section:
- restores and engineers the `EXT_SOURCE_*` features from the original dataset,
- uses stratified cross-validation,
- compares multiple classifiers on a sampled subset of the training data to reduce runtime,
- performs logistic regression grid search on the full training set,
- and evaluates both ROC-AUC and PR-AUC on the untouched holdout test set.


```python
# Build the improved model from the original dataset so that high-signal variables are retained
model_df = df.copy()

# Recreate and preserve the external risk score features instead of dropping them
ext_cols = [c for c in ["EXT_SOURCE_1", "EXT_SOURCE_2", "EXT_SOURCE_3"] if c in model_df.columns]
if len(ext_cols) > 0:
    model_df["EXT_SOURCE_MEAN"] = model_df[ext_cols].mean(axis=1)
    model_df["EXT_SOURCE_STD"] = model_df[ext_cols].std(axis=1)

for c in ext_cols:
    model_df[f"{c}_MISSING"] = model_df[c].isna().astype(int)

# Rebuild engineered ratio features from the original data
model_df["DAYS_EMPLOYED_ANOM"] = (model_df["DAYS_EMPLOYED"] == 365243).astype(int)
model_df["DAYS_EMPLOYED"] = model_df["DAYS_EMPLOYED"].replace(365243, np.nan)
model_df["YEARS_BIRTH"] = -model_df["DAYS_BIRTH"] / 365
model_df["YEARS_EMPLOYED"] = -model_df["DAYS_EMPLOYED"] / 365

model_df["CREDIT_INCOME_RATIO"] = model_df["AMT_CREDIT"] / model_df["AMT_INCOME_TOTAL"].replace(0, np.nan)
model_df["ANNUITY_INCOME_RATIO"] = model_df["AMT_ANNUITY"] / model_df["AMT_INCOME_TOTAL"].replace(0, np.nan)
model_df["GOODS_CREDIT_RATIO"] = model_df["AMT_GOODS_PRICE"] / model_df["AMT_CREDIT"].replace(0, np.nan)
model_df["CREDIT_ANNUITY_RATIO"] = model_df["AMT_CREDIT"] / model_df["AMT_ANNUITY"].replace(0, np.nan)

# Remove identifier column if present
for col in ["SK_ID_CURR"]:
    if col in model_df.columns:
        model_df = model_df.drop(columns=col)

X_improved = model_df.drop(columns=["TARGET"])
y_improved = model_df["TARGET"]

numeric_cols_improved = X_improved.select_dtypes(exclude=["object"]).columns.tolist()
categorical_cols_improved = X_improved.select_dtypes(include=["object"]).columns.tolist()

print("Improved modeling dataframe shape:", model_df.shape)
print("Numeric features:", len(numeric_cols_improved))
print("Categorical features:", len(categorical_cols_improved))
```

```python
# Train/test split for improved modeling workflow
X_train_imp, X_test_imp, y_train_imp, y_test_imp = train_test_split(
    X_improved, y_improved, test_size=0.20, stratify=y_improved, random_state=42
)
```

### Efficient model comparison strategy

Because the training dataset is large, I use a representative sample of the training data only for the cross-validated model-comparison stage. This reduces runtime while still allowing me to compare candidate models. After selecting the most promising model family, I tune and fit the final model on the full training dataset and then evaluate it on the untouched holdout test set.

```python
# Create a smaller sample of the training data for faster model comparison
sample_n = min(100000, len(X_train_imp))

X_train_cmp = X_train_imp.sample(n=sample_n, random_state=42)
y_train_cmp = y_train_imp.loc[X_train_cmp.index]

print("Full training shape:", X_train_imp.shape)
print("Comparison sample shape:", X_train_cmp.shape)
```

```python
# Model comparison on a sampled subset of the training data

numeric_transformer_imp = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer_imp = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor_imp = ColumnTransformer(transformers=[
    ("num", numeric_transformer_imp, numeric_cols_improved),
    ("cat", categorical_transformer_imp, categorical_cols_improved)
])

# Use 3-fold CV to keep runtime reasonable in standard Colab
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

models = {
    "LogisticRegression": LogisticRegression(
        max_iter=1000,
        class_weight="balanced",
        random_state=42
    ),
    "RandomForest": RandomForestClassifier(
        n_estimators=50,
        max_depth=10,
        min_samples_leaf=5,
        class_weight="balanced",
        n_jobs=-1,
        random_state=42
    )
}

comparison_rows = []

for model_name, model in models.items():
    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor_imp),
        ("model", model)
    ])

    scores = cross_validate(
        pipe,
        X_train_cmp,
        y_train_cmp,
        cv=cv,
        scoring={"roc_auc": "roc_auc", "pr_auc": "average_precision"},
        n_jobs=-1
    )

    comparison_rows.append({
        "Model": model_name,
        "CV ROC-AUC Mean": scores["test_roc_auc"].mean(),
        "CV ROC-AUC Std": scores["test_roc_auc"].std(),
        "CV PR-AUC Mean": scores["test_pr_auc"].mean(),
        "CV PR-AUC Std": scores["test_pr_auc"].std()
    })

results_df = pd.DataFrame(comparison_rows).sort_values("CV ROC-AUC Mean", ascending=False)
results_df
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.barplot(data=results_df, x="Model", y="CV ROC-AUC Mean", ax=axes[0])
axes[0].set_title("Cross-Validated ROC-AUC by Model", fontsize=12)
axes[0].set_xlabel("Model")
axes[0].set_ylabel("Mean CV ROC-AUC")
for i, v in enumerate(results_df["CV ROC-AUC Mean"]):
    axes[0].text(i, v + 0.002, f"{v:.3f}", ha="center", fontsize=9)

sns.barplot(data=results_df, x="Model", y="CV PR-AUC Mean", ax=axes[1])
axes[1].set_title("Cross-Validated PR-AUC by Model", fontsize=12)
axes[1].set_xlabel("Model")
axes[1].set_ylabel("Mean CV PR-AUC")
for i, v in enumerate(results_df["CV PR-AUC Mean"]):
    axes[1].text(i, v + 0.002, f"{v:.3f}", ha="center", fontsize=9)

fig.suptitle("Logistic Regression Delivered the Strongest Validation Performance", fontsize=14, y=1.03)
plt.tight_layout()
plt.show()
```

### Why ROC-AUC and PR-AUC?

This is an imbalanced classification problem, so accuracy alone would be misleading. ROC-AUC is appropriate because it measures ranking quality across thresholds, while PR-AUC is useful because it focuses more directly on performance for the minority class.

```python
log_pipe = Pipeline(steps=[
    ("preprocessor", preprocessor_imp),
    ("model", LogisticRegression(max_iter=1000, random_state=42))
])

param_grid = {
    "model__C": [0.1, 1.0, 10.0],
    "model__class_weight": [None, "balanced"],
    "model__solver": ["liblinear"]
}

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

grid = GridSearchCV(
    estimator=log_pipe,
    param_grid=param_grid,
    scoring="roc_auc",
    cv=cv,
    n_jobs=-1,
    verbose=1,
    refit=True
)

# Important: final tuning is done on the full improved training set
grid.fit(X_train_imp, y_train_imp)

print("Best parameters:", grid.best_params_)
print("Best cross-validated ROC-AUC:", round(grid.best_score_, 4))
```

```python
# GridSearchCV refits the best estimator on the full training set by default
final_model = grid.best_estimator_

y_prob_imp = final_model.predict_proba(X_test_imp)[:, 1]
y_pred_imp = (y_prob_imp >= 0.50).astype(int)

roc_auc_imp = roc_auc_score(y_test_imp, y_prob_imp)
pr_auc_imp = average_precision_score(y_test_imp, y_prob_imp)

print("Improved holdout ROC-AUC:", round(roc_auc_imp, 4))
print("Improved holdout PR-AUC:", round(pr_auc_imp, 4))
print()
print(classification_report(y_test_imp, y_pred_imp))

RocCurveDisplay.from_predictions(y_test_imp, y_prob_imp)
plt.title("ROC Curve - Tuned Logistic Regression")
plt.tight_layout()
plt.show()

PrecisionRecallDisplay.from_predictions(y_test_imp, y_prob_imp)
plt.title("Precision-Recall Curve - Tuned Logistic Regression")
plt.tight_layout()
plt.show()

ConfusionMatrixDisplay.from_predictions(y_test_imp, y_pred_imp)
plt.title("Confusion Matrix at 0.50 Threshold")
plt.tight_layout()
plt.show()
```
